```python
import re
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime
import numpy as np
```

# CROCv1-50um

### 20240307_104956

```python
with open('/Users/westonschwartz/Desktop/Detector_Work/wafer-test-logs/CROCv1-50um/wafer_N6Y250-16G4_20240307_104956.log', 'r') as file:
    log_content = file.readlines()

pattern = r'(\d{2}:\d{2}:\d{2},\d{3}).*Contact resistance \(analog\): (\d+\.\d+) ohm'
time_pattern = r'(\d{2}:\d{2}:\d{2},\d{3}).*Performing a cleaning procedure on the needles!'

timestamps = []
resistances = []

for line in log_content:
    match = re.search(pattern, line)
    if match:
        timestamp = match.group(1)
        resistance = float(match.group(2))  # Convert resistance to float
        timestamps.append(timestamp)
        resistances.append(resistance)
        
times = []

for i in log_content:
    good = re.search(time_pattern, i)
    if good:
        time = good.group(1)
        times.append(time)

# Fixed date for all timestamps
fixed_date = '2024-04-13'
dates = [datetime.strptime(f'{fixed_date} {ts}', '%Y-%m-%d %H:%M:%S,%f') for ts in timestamps]

# Plot
plt.figure(figsize=(12, 6))
plt.plot(dates, resistances, marker='o', linestyle='-')

# Add vertical lines for each time in the times list
for time in times:
    time_dt = datetime.strptime(f'{fixed_date} {time}', '%Y-%m-%d %H:%M:%S,%f')
    plt.axvline(x=time_dt, color='r', linestyle='--')
    plt.text(time_dt, plt.gca().get_ylim()[1], time, ha='center', va='bottom', color='r')

# Format the x-axis
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
plt.gca().xaxis.set_major_locator(mdates.HourLocator(interval=1))
plt.gcf().autofmt_xdate()

# Set labels and title
plt.xlabel('Time')
plt.ylabel('Resistance (ohm)')
plt.title('Analog Resistance vs. Time 104956', y=1.05)

# Get 16 evenly spaced indices from the dates list
indices = np.linspace(0, len(dates) - 1, num=16, dtype=int)

# Use the indices to get the corresponding dates and set them as tick locations and labels
plt.xticks([dates[i] for i in indices], [dates[i].strftime('%H:%M:%S') for i in indices])

# Show plot
plt.show()
```
